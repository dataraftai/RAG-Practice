```python
## Create a simple txt file

import os
os.makedirs("../data/text_files",exist_ok=True)
```

```python
sample_texts = {
    "../data/text_files/python_intro.txt":"""Python is a high-level, interpreted programming language known for its clear syntax and exceptional readability. Created by Guido van Rossum and first released in 1991, it has become one of the most popular languages in the world. It supports multiple programming paradigms, including object-oriented, procedural, and functional programming.Key Features of PythonSimple Syntax: Reads similarly to plain English.Dynamically Typed: Variables do not require explicit type declaration.Interpreted: Code executes line-by-line immediately without prior compilation.Cross-Platform: Runs seamlessly on Windows, macOS, Linux, and Raspberry Pi.Large Ecosystem: Packed with massive standard and third-party libraries.Core Concepts1. Your First ProgramPython uses the print() function to display text to the console. Comments are created using the # symbol.python# This is a comment and will be ignored by Python
print("Hello, World!"). 
""",
    "../data/text_files/machine_learning.txt" : """Machine learning (ML) is a subset of artificial intelligence (AI) that enables computers to learn from data and make predictions without being explicitly programmed. Instead of writing hard-coded instructions, ML models identify patterns in data to autonomously solve tasks ranging from medical diagnostics to spam filtering."""

}

for filepath,content in sample_texts.items():
    with open(filepath,"w",encoding="utf-8") as f:
        f.write(content)

print("Sample text files created")
```

```python
### how to read those files
# TextLoader

from langchain_community.document_loaders import TextLoader
loader = TextLoader("../data/text_files/python_intro.txt",encoding="utf-8")
document = loader.load()
print(document)
```

```python
# Directory Loader

from langchain_community.document_loaders import DirectoryLoader

dir_Loader = DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",  ## pattern to match files
    loader_kwargs={"encoding": "utf-8"},
    loader_cls=TextLoader,
    show_progress=False
)

documents = dir_Loader.load()
documents
```
